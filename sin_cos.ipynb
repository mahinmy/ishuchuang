```python
import pandas as pd
all_data=pd.read_csv('all_data.csv',index_col=0)
all_data
```

```python
from math import sin,cos,pi
all_data['横向风速']=all_data[['汕尾风速','汕尾风向']].apply(lambda x:x['汕尾风速']*sin(pi/180*x['汕尾风向']),axis=1)
all_data['纵向风速']=all_data[['汕尾风速','汕尾风向']].apply(lambda x:x['汕尾风速']*cos(pi/180*x['汕尾风向']),axis=1)
all_data
```
