```python
three_minus_zero_dt=pd.DataFrame({'汕尾风速':[],'风速1':[],'风速2':[],'风速3':[],'风速4':[]})
for i in range(0,2920,8):
    minus=pd.DataFrame((all_data.iloc[i+1,1:6]-all_data.iloc[i,1:6]).astype('float32'))
    three_minus_zero_dt=pd.concat([three_minus_zero_dt,minus.T],ignore_index=True)
three_minus_zero_dt
```

```python
minus_list=[]
for k in range(0,8):
    minus_dt=pd.DataFrame({'汕尾风速':[],'风速1':[],'风速2':[],'风速3':[],'风速4':[]})
    for i in range(0+k,2919,8):
        minus=pd.DataFrame((all_data.iloc[i+1,1:6]-all_data.iloc[i,1:6]).astype('float32'))
        minus_dt=pd.concat([minus_dt,minus.T],ignore_index=True)
    minus_list.append(minus_dt)
for m in minus_list:
    print(m)
```
