```python
from repeat_work import read_one_month
from repeat_work import read_one_year
import pandas as pd
from repeat_work import random_search
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
data=pd.read_csv('data1993/1993-1月.csv',encoding='gbk')
for month in range(2,13):
    file_name=f'data1993/1993-{month}月.csv'
    data=pd.concat([data,pd.read_csv(file_name,encoding='gbk')],ignore_index=True)
for month in range(1,13):
    file_name=f'data1994/1994-{month}月.csv'
    data=pd.concat([data,pd.read_csv(file_name,encoding='gbk')],ignore_index=True)
for month in range(1,13):
    file_name=f'data1995/1995-{month}月.csv'
    data=pd.concat([data,pd.read_csv(file_name,encoding='gbk')],ignore_index=True)
from math import sin,cos,pi
data['横向风速']=data[['汕尾风速','汕尾风向']].apply(lambda x:x['汕尾风速']*sin(pi/180*x['汕尾风向']),axis=1)
data['纵向风速']=data[['汕尾风速','汕尾风向']].apply(lambda x:x['汕尾风速']*cos(pi/180*x['汕尾风向']),axis=1)
data.to_csv('all_data.csv',index=False,header=True)
```

```python
MIN=data['横向风速'].min()
MAX=data['横向风速'].max()
data['横向风速scaled']=data['横向风速'].apply(lambda x:2*(x-MIN)/(MAX-MIN)-1)
MIN=data['纵向风速'].min()
MAX=data['纵向风速'].max()
data['纵向风速scaled']=data['纵向风速'].apply(lambda x:2*(x-MIN)/(MAX-MIN)-1)
data
```

```python
X_all=data[['横向风速scaled','纵向风速scaled']]
y_all=data[['汕尾波高']]
X_all
```

```python
def create_data(days=5):
    X_data=[]
    Y_data=np.array([])
    for i in range(8760-days):
        X_data.append(data.iloc[i:i+days,[28,29,11]].values)
        Y_data=np.append(Y_data,data.iloc[i+days,11])
    return np.array(X_data),Y_data
X,y=create_data(10)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train=X[0:7000]
X_valid=X[7000:7875]
X_test=X[7875:]
y_train=y[0:7000]
y_valid=y[7000:7875]
y_test=y[7875:]
y_test.shape
```

```python
from repeat_work import random_search
random_search_cv = random_search(X_train,y_train,X_valid,y_valid,cv=10)
```

```python
print(random_search_cv.best_params_)
print(random_search_cv.best_score_)
print(random_search_cv.best_estimator_)
```

上面的在random search中训练集分成了10份进行交叉验证得到的最优模型的结果。

```python
model = random_search_cv.best_estimator_.model
model.evaluate(X_test,y_test)
```
